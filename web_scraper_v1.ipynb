```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver import Chrome
import pandas as pd
import time
from random import randint
import requests
```

```python
url = 'https://www.brookings.edu/events/upcoming/'
response = requests.get(url)
```

```python
page = response.text
soup = BeautifulSoup(page)
soup
```

```python
driver = webdriver.Chrome(executable_path='/Users/yawagipong/Downloads/chromedriver_mac_arm64/chromedriver')
driver.get(url)
results = []
content = driver.page_source
soup = BeautifulSoup(content)
```

```python
for element in soup.findAll(attrs = {'class': 'title'}):
  name = element.find(element)
  if name not in results:
    results.append(name.text)
```
